```python
def count_chat_tokens(messages, model="gpt-4o-mini"):
    enc = tiktoken.encoding_for_model(model)
    tokens = 0
    for msg in messages:
        tokens += len(enc.encode(msg["content"]))
    return tokens
```

```python
from anthropic_tokenizer import count_tokens, ApiTokenizer

tokenizer = ApiTokenizer()
tokens = tokenizer.count_tokens("Hello, world!")
print(tokens)

```

```python
def count_claude_tokens(text):
  tokenizer = ApiTokenizer()
  tokens = tokenizer.count_tokens(text)
  return tokens
```

```python
def compare_token_price(prompt):
  # Define approximate pricing per million tokens (input tokens for simplicity)
  # Claude 3 Sonnet (example pricing: $3.00/1M input tokens)
  claude_cost_per_million_tokens = 3.00
  # GPT-4o mini (example pricing: $0.15/1M input tokens)
  gpt4o_mini_cost_per_million_tokens = 0.15

  # Count tokens for Claude
  claude_tokens = count_claude_tokens(prompt)
  claude_cost = (claude_tokens / 1_000_000) * claude_cost_per_million_tokens

  # Count tokens for GPT-4o-mini
  gpt4o_mini_tokens = count_chat_gpt_token(prompt, model="gpt-4o-mini")
  gpt4o_mini_cost = (gpt4o_mini_tokens / 1_000_000) * gpt4o_mini_cost_per_million_tokens

  print(f"Prompt: '{prompt[:50]}...'\n")
  print(f"--- Claude 3 Sonnet ---")
  print(f"Tokens: {claude_tokens}")
  print(f"Estimated cost: ${claude_cost:.6f}\n")

  print(f"--- GPT-4o Mini ---")
  print(f"Tokens: {gpt4o_mini_tokens}")
  print(f"Estimated cost: ${gpt4o_mini_cost:.6f}\n")

  if claude_cost < gpt4o_mini_cost:
    print(f"Claude 3 Sonnet is estimated to be cheaper for this prompt.")
  elif gpt4o_mini_cost < claude_cost:
    print(f"GPT-4o Mini is estimated to be cheaper for this prompt.")
  else:
    print(f"Both models have similar estimated costs for this prompt.")
```

```python
prompt="What is quantam entangalment ?"
compare_token_price(prompt)
```

```python
prompts_varied = [
    "Describe the color blue to someone who has never seen it.",
    "Summarize the plot of the novel '1984' in exactly fifty words.",
    "Create a short, whimsical poem about a talking squirrel who loves to knit sweaters.",
    "Explain the working principle of a quantum computer, highlighting its potential applications and current limitations.",
    "List five distinct challenges faced by renewable energy adoption and suggest a practical solution for each.",
    "Write a dialogue between a grumpy old wizard and a curious young dragon about the best way to brew tea.",
    "What is the historical significance of the Magna Carta?",
    "Generate a creative writing prompt involving a forgotten celestial map and a journey to a hidden constellation.",
    "Outline a step-by-step guide for making authentic Italian lasagna from scratch.",
    "Analyze the socio-economic impact of the Industrial Revolution on 19th-century European society, considering both positive and negative aspects."
]
```

```python
for prompt in prompts_varied:
  compare_token_price(prompt)
  print("\n\n")
```

```python

import pandas as pd

def get_token_comparison_data(prompt):
  """Calculates token counts and estimated costs for Claude and GPT-4o Mini and returns them as a dictionary."""
  claude_cost_per_million_tokens = 3.00
  gpt4o_mini_cost_per_million_tokens = 0.15

  claude_tokens = count_claude_tokens(prompt)
  claude_cost = (claude_tokens / 1_000_000) * claude_cost_per_million_tokens

  gpt4o_mini_tokens = count_chat_gpt_token(prompt, model="gpt-4o-mini")
  gpt4o_mini_cost = (gpt4o_mini_tokens / 1_000_000) * gpt4o_mini_cost_per_million_tokens

  return {
      "Prompt": prompt,
      "Claude_Tokens": claude_tokens,
      "Claude_Cost_USD": claude_cost,
      "GPT4o_Mini_Tokens": gpt4o_mini_tokens,
      "GPT4o_Mini_Cost_USD": gpt4o_mini_cost,
  }

def export_comparison_to_excel(comparison_data_list, filename="token_comparison.xlsx"):
  """Exports a list of token comparison dictionaries to an Excel file."""
  df = pd.DataFrame(comparison_data_list)
  df.to_excel(filename, index=False)
  print(f"Successfully exported token comparison data to '{filename}'")

# --- Example Usage ---
# Collect data for all prompts in prompts_varied
all_comparison_results = []
for prompt in prompts_varied:
  result = get_token_comparison_data(prompt)
  all_comparison_results.append(result)

# Export the collected data to an Excel file
df=pd.DataFrame(all_comparison_results)
df.to_csv("token_comparison.csv")
```
