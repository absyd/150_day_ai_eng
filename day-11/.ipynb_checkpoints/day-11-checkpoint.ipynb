### Vectorised ops — the no-loop way


```python
# Element-wise arithmetic (no loop needed)
scores = np.array([85,92,78,96])
scores.sum()
```

```python
normalised = (scores - scores.mean()) / scores.std()
normalised
```

```python
# Boolean masking (replaces if-loops)
high_scores = scores[scores > 90]  # [92, 96]
high_scores
```

### Universal functions (ufuncs) — compiled C, extremely fast


```python
# Universal functions (ufuncs) — compiled C, extremely fast
np.sqrt(scores)
```

# Broadcasting — the shape-matching superpower

- Broadcasting lets arrays of different shapes operate together without copying data. NumPy expands dimensions implicitly by the rules below.



### The 3 broadcasting rules
1. Pad dimensions on the left . If shapes have different number of dims, prefix 1s. Shape (3,) becomes (1, 3).
2. Expand size-1 dimensions
Any axis of size 1 is stretched to match the other array's size along that axis.
3. Sizes must match or be 1
If neither axis is 1 and sizes differ, you get a ValueError. No silent surprises.


Shape compatibility chart
A shape	B shape	Result shape	Works?
(3, 4)	(4,)	(3, 4)	yes
(3, 1)	(1, 4)	(3, 4)	yes
(3, 4)	(3,)	—	error
(1, 100, 1)	(32, 1, 768)	(32, 100, 768)	yes
(3, 4)	(3, 5)	—	error

```python
# Pattern 1: Normalise each row independently
X = rng.random((1000, 20))  # 1000 samples, 20 features
mean = X.mean(axis=0)      # shape (20,)
std  = X.std(axis=0)       # shape (20,)
X_norm = (X - mean) / std    # broadcast: (1000,20)-(20,) → works!
X_norm
```

```python
mean.shape
```

```python
# Pattern 2: Pairwise distance (no loop)
a = rng.random((100, 3))     # 100 points in 3D
b = rng.random((200, 3))     # 200 points in 3D
# a[:,None,:] → (100,1,3)  b[None,:,:] → (1,200,3)
diffs = a[:, None, :] - b[None, :, :]  # (100,200,3)
dists = np.sqrt((diffs**2).sum(axis=-1)) # (100,200)
```

# Linear algebra — matmul, dot, norm, SVD

- NumPy delegates to BLAS/LAPACK — the same routines used in MATLAB. Understanding what these ops do geometrically is what interviewers test.



| Operation  | Use Case              | Batched? | Note                                    |
|------------|-----------------------|----------|-----------------------------------------|
| `np.dot`   | Dot product, matrix mult | no     | Avoid for **2D+ arrays**; semantics get confusing. |
| `@`/ `matmul` | Matrix × matrix, matrix × vector | yes  | **Preferred**. Works on stacks of matrices. |
| `np.einsum`| Any contraction you can describe | yes | Most expressive but slower if not optimized. |

```python
A = rng.random((3, 4))
A
```

```python
B =rng.random((4,5))
B
```

```python
C=A@B
C
```

```python
# Batched matmul — e.g. attention heads
Q = rng.random((8, 64, 32))  # 8 heads, 64 tokens, 32 dim
K = rng.random((8, 32, 64))
scores = Q @ K               # (8, 64, 64) — batch matmul!
print("Q",Q)
print("K",K)
print("Scores",scores)
```

```python
# einsum — trace of matrix product
np.einsum('ij,ji->', A, A.T)  # scalar: trace(A Aᵀ)
```
