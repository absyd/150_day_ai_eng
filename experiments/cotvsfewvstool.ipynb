# 🧠 0. Big Picture Mental Model

You have **3 fundamentally different ways to make an LLM smarter**:

| Technique                  | What it does                                 |
| -------------------------- | -------------------------------------------- |
| **Few-Shot**               | Teaches pattern via examples                 |
| **Chain-of-Thought (CoT)** | Improves reasoning via step-by-step thinking |
| **Tool Use**               | Extends capability via external systems      |

👉 Think of it like:

* Few-shot = “show examples”
* CoT = “think step by step”
* Tools = “use external power”

---

# 🔬 1. Few-Shot Prompting (Recap → Advanced Framing)

## What it really is

Few-shot = **implicit function definition via examples**

You are telling the model:

> “Here are input-output mappings. Learn the transformation.”

---

## When Few-Shot works BEST

✅ Classification
✅ Formatting / structured output
✅ Style imitation
✅ Translation
✅ Simple transformations

---

## When it FAILS

❌ Multi-step reasoning
❌ Math-heavy problems
❌ Logical deduction chains

---

## Example

```text
Input: "I love this" → Positive
Input: "This is terrible" → Negative

Input: "Not bad but could be better" →
```

👉 Model copies pattern

---

# 🧠 2. Chain-of-Thought (CoT)

## What it is

Chain-of-Thought = forcing the model to **explicitly reason step-by-step**

---

## Basic Prompt

```text
Solve step by step:

If a train travels 60 km in 1 hour, how far in 3.5 hours?
```

---

## Why it works (DEEP INSIGHT)

Without CoT:

* Model tries to jump → more errors

With CoT:

* Model decomposes problem → higher accuracy

👉 This activates **reasoning pathways**

---

## 🔬 Types of CoT

### 1. Zero-shot CoT

```text
"Think step by step"
```

Simple but powerful

---

### 2. Few-shot CoT (VERY POWERFUL)

```text
Q: 2+3?
A: Let's think step by step: 2+3 = 5

Q: 4×6?
A: Let's think step by step: 4×6 = 24

Q: 7×8?
A:
```

👉 You combine:

* Examples (few-shot)
* Reasoning (CoT)

---

### 3. Hidden CoT (Production)

You ask:

```text
Give only final answer.
```

But internally model reasons.

---

## When CoT works BEST

✅ Math problems
✅ Logical puzzles
✅ Multi-step reasoning
✅ Algorithmic thinking
✅ Debugging code

---

## When CoT is OVERKILL

❌ Simple classification
❌ Formatting
❌ Short answers

---

# 🧰 3. Tool Use (Function Calling / Agents)

## What it is

Tool use = **LLM calls external systems**

Examples:

* Calculator
* Database
* API
* Code execution
* Search

---

## Example

User:

```text
What is 234 * 567?
```

Instead of guessing:

👉 LLM calls calculator tool
👉 Returns exact answer

---

## 🔬 Why Tool Use is CRITICAL

LLMs are:

❌ Not reliable at:

* Arithmetic
* Real-time data
* Long computations

👉 Tools fix this

---

## Common Tools

| Tool             | Purpose         |
| ---------------- | --------------- |
| Calculator       | Math            |
| Code interpreter | Complex logic   |
| Retrieval (RAG)  | Knowledge       |
| APIs             | Real-world data |

---

## When Tool Use is BEST

✅ High accuracy required
✅ External data needed
✅ Deterministic tasks
✅ Long computations

---

## When NOT to use tools

❌ Simple reasoning
❌ Cheap tasks
❌ No external dependency

---

# ⚔️ 4. Direct Comparison (IMPORTANT)

| Feature     | Few-Shot         | CoT           | Tool Use             |
| ----------- | ---------------- | ------------- | -------------------- |
| Purpose     | Pattern learning | Reasoning     | Capability extension |
| Strength    | Consistency      | Accuracy      | Reliability          |
| Weakness    | Poor reasoning   | Slow, verbose | Complex setup        |
| Cost        | Medium           | High tokens   | High infra           |
| Determinism | Medium           | Low           | High                 |

---

# 🧠 5. Decision Framework (REAL-WORLD)

Use this in your projects:

---

## Step 1: Is it pattern-based?

👉 YES → Use **Few-Shot**

---

## Step 2: Is it multi-step reasoning?

👉 YES → Use **CoT**

---

## Step 3: Need external accuracy/data?

👉 YES → Use **Tool Use**

---

## 🔥 Combined Strategy (PRO LEVEL)

Most real systems use ALL THREE.

---

## Example: AI Coding Assistant

User:
“Find bug in this code and fix it”

### Pipeline:

1. Few-shot → show bug-fix examples
2. CoT → reason through code
3. Tool → run code / test output

👉 This is **production-grade architecture**

# 🧪 6. Hybrid Patterns (VERY IMPORTANT)

---

## 🔹 Few-Shot + CoT

Best for:

* Complex reasoning with structure

Example:

```text
Q: If 2 apples cost $4, what is price of 5 apples?
A: Let's think step by step:
2 apples → $4
1 apple → $2
5 apples → $10

Q: If 3 pens cost $6, what is price of 7 pens?
A:
```

---

## 🔹 CoT + Tool Use

Best for:

* Math + execution

Flow:

1. Reason
2. Call calculator

---

## 🔹 Few-Shot + Tool Use

Best for:

* API formatting
* Function calling
