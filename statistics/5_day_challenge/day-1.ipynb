```python
import numpy as np
import pandas as pd
from scipy import stats

data = [10, 20, 20, 30, 100]

# Mean, Median, Mode
print("Mean:",np.mean(data))
print("Median:",np.median(data))
print("Mode:",list(stats.mode(data))[0])

# Standard deviation
print("Standard deviation:",np.std(data))

# Skewness & Kurtosis
print("Skewness:",stats.skew(data))
print("Kurtosis:",stats.kurtosis(data))

# Z-score
print("Z-score:",stats.zscore(data))
```

```python
sns.scatterplot(data=titanic, x='age', y='fare')
plt.xlabel('Age')
plt.ylabel('Fare')
# show in red those who class= 1 and label if alive or not
mask = titanic['pclass'] == 1
plt.scatter(titanic['age'][mask], titanic['fare'][mask], c='r', label = 'Survived' if titanic['survived'][mask].values[0] else 'Died')
plt.show()
```
